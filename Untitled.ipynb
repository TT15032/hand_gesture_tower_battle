```python
# splitエラー回避策
# https://learning.unity3d.jp/7557/
import onnx
# Splitエラー対策
def fix_split_ops(model):
    for i in range(len(model.graph.node)):
        node = model.graph.node[i]
        if node.op_type != 'Split':
            continue
        
        split = (1, ) * len(node.output)
        new_node = onnx.helper.make_node('Split', node.input, node.output, split = split)
        model.graph.node.insert(i, new_node)
        model.graph.node.remove(node)
# onnx_model is an in-memory ModelProto
onnx_model = onnx.load('runs/segment/train14/weights/best.onnx')
fix_split_ops(onnx_model)
onnx.save(onnx_model, 'yolov8n-2.onnx')
```

```python
from ultralytics import YOLO
import onnx
def fix_onnx_resize(model):
    for i in range(len(model.graph.node)):
        node = model.graph.node[i]
        if node.op_type != 'Resize':
            continue
        
        new_node = onnx.helper.make_node(
                'Resize',
                inputs=node.input,
                outputs=node.output,
                name=node.name,
                coordinate_transformation_mode='half_pixel',  # Instead of pytorch_half_pixel, unsupported by Tensorflow
                mode='linear',
            )
        model.graph.node.insert(i, new_node)
        model.graph.node.remove(node)
model = YOLO('runs/segment/train14/weights/best.pt')
model.export(format='onnx', opset=12,simplify=True, dynamic=False)
```

```python
from ultralytics import YOLO
import onnx
from onnx import helper, numpy_helper

def fix_resize_nodes(model):
    for i, node in enumerate(model.graph.node):
        if node.op_type == 'Resize':
            attrs = {
                'mode': 'nearest',  # or 'linear'
                'coordinate_transformation_mode': 'half_pixel',
                'nearest_mode': 'floor'
            }

            # 再構築された Resize ノードを作成
            new_node = helper.make_node(
                'Resize',
                inputs=node.input,
                outputs=node.output,
                name=node.name,
                **attrs
            )

            model.graph.node.remove(node)
            model.graph.node.insert(i, new_node)

def fix_div_rank_issue(model):
    # ValueInfoProtoが不足している場合に追加
    needed_inputs = set()
    for node in model.graph.node:
        if node.op_type == 'Div':
            for input_name in node.input:
                needed_inputs.add(input_name)

    existing_info = {vi.name for vi in model.graph.value_info}
    for input_name in needed_inputs - existing_info:
        # 仮の形状情報を追加 (4D: NCHW)
        vi = helper.make_tensor_value_info(input_name, onnx.TensorProto.FLOAT, [1, 32, 160, 160])
        model.graph.value_info.append(vi)

# エクスポート
model = YOLO('runs/segment/train14/weights/best.pt')
model.export(format='onnx', opset=12, simplify=True)

# ONNX モデル読み込みと修正
onnx_model = onnx.load('runs/segment/train14/weights/best.onnx')
fix_resize_nodes(onnx_model)
fix_div_rank_issue(onnx_model)
onnx.save(onnx_model, 'yolov8n-barracuda-fixed11.onnx')
```
